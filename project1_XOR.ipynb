```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
```

```python
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Class 1')
plt.xlabel('x₁')
plt.ylabel('x₂')
plt.title('XOR Dataset')
plt.legend()
plt.show()
```

```python
mlp_clf = MLPClassifier(hidden_layer_sizes=(4,), activation='tanh', solver='lbfgs', random_state=42)
mlp_clf.fit(X, y)
```

```python
mlp_pred = mlp_clf.predict(X)
print("MLP Predictions:", mlp_pred)
print("Actual labels:", y)
```

```python
mlp_acc = accuracy_score(y, mlp_pred)
print("MLP Accuracy:", mlp_acc)
```

```python
svm_clf = SVC(kernel='rbf', C=1.0, gamma='auto')
svm_clf.fit(X, y)
```

```python
svm_pred = svm_clf.predict(X)
print("SVM Predictions:", svm_pred)
print("Actual labels:", y)
```

```python
svm_acc = accuracy_score(y, svm_pred)
print("SVM Accuracy:", svm_acc)
```

```python
def plot_decision_boundary(model, title):
    xx, yy = np.meshgrid(np.linspace(-0.5, 1.5, 200), np.linspace(-0.5, 1.5, 200))
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    Z = model.predict(grid_points)
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='red', edgecolor='k', label='Class 0')
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', edgecolor='k', label='Class 1')
    plt.xlabel('x₁')
    plt.ylabel('x₂')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
plot_decision_boundary(mlp_clf, "MLP Decision Boundary")
```

```python
plot_decision_boundary(svm_clf, "SVM Decision Boundary")
```
